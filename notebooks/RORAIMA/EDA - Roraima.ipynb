## Venezuela

```python
export_df = pd.read_csv('/home/breno/proj/roraima-energy-social/VENEZUELA/data/3301.csv', index_col=0)
import_df = pd.read_csv('/home/breno/proj/roraima-energy-social/VENEZUELA/data/3533.csv', index_col=0)
```

```python
x = import_df['date']

trace1 = go.Scatter(x = x, y = import_df['value'], name = 'Imports')
trace2 = go.Scatter(x = x, y = export_df['value'], name = 'Exports')

layout = go.Layout(title = 'Venezuelan Imports and Exports - BCB', xaxis = dict(type = 'date'),
                   yaxis = dict(title = 'USD in million'), template='plotly_dark')
fig = go.Figure(data=[trace1,trace2], layout=layout)
pyo.iplot(fig, config=dict(showLink = False))
```
