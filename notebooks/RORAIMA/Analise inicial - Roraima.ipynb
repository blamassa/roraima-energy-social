```python
import plotly.offline as pyo
from plotly import graph_objs as go

pyo.init_notebook_mode(connected=True)

f = open('index.md','w')

to_write = '''
# SUMÁRIO DOS GRÁFICOS

Os dados foram retirados da biblioteca BETS (Brazilian Economic Time Series) do R, elaborado por uma equipe da FGV.
Clique no código para visualizar o gráfico.

| index | code | descrição | unidade | periodicidade | início | final | fonte |
| --- | ---  | ---       | ---     |  ----         | ---    | ---   | ---   | \n'''

f.write(to_write)
f.close()


for i in range(roraima.shape[0]):

    df = pd.read_csv('/home/breno/proj/BETS/time_series/'+str(roraima['code'][i])+'.csv')

    y = df['value']
    x = df['date']


    # plotar o grafico
    trace = go.Scatter(x = x, y = y)
    layout = go.Layout(title = roraima['description'][i],
                       font = dict(size=20),
                      xaxis = dict(title = 'Fonte: '+roraima['source'][i], 
                                   fixedrange=True),
                      yaxis = dict(title = roraima['unit'][i], 
                                   fixedrange=True))#                  hovermode = 'closest')

    fig = go.Figure([trace], layout=layout)

    config = dict(showLink = False, displayModeBar = False)

    pyo.plot(fig, config = config, filename='/home/breno/proj/BETS/graphs/'+str(roraima['code'][i])+'.html')
    
    # Salvar linha no md
    f = open('index.md','a')
    
    code = str(roraima['code'][i])
    descricao = roraima['description'][i]
    unidade = roraima['unit'][i]
    periodo = roraima['periodicity'][i]
    inicio = roraima['start'][i]
    final = roraima['last_value'][i]
    fonte = roraima['source'][i]
    
    to_write = '| '+ str(i) +' | '+code+' | ['+descricao+'](./graphs/'+code+'.html) | '+unidade+' | '+periodo+' | '+inicio+' | '+final+' | '+fonte+' | \n'
    
    f.write(to_write)
    f.close()
```
